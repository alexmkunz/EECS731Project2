```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

shksprData = pd.read_csv('../data/Shakespeare_data.csv')
shksprData.head()
```

```python
# -----------------------------------------------------------------
# Remove useless data
# -----------------------------------------------------------------

# Delete pointless Dataline column
del shksprData['Dataline']
# Delete PlayerLine since we will not use it in our algorithm
del shksprData['PlayerLine']
# Delete rows that include one or more NaN values
shksprData = shksprData.dropna()

# -----------------------------------------------------------------

shksprData.head()
```

```python
# -----------------------------------------------------------------
# Make categorical features useable with encoding
# -----------------------------------------------------------------

le = LabelEncoder()
leFeatures = ['Play', 'PlayerLinenumber', 'ActSceneLine']
shksprData[leFeatures] = shksprData[leFeatures].apply(lambda col: le.fit_transform(col))

# -----------------------------------------------------------------

shksprData.head()
```
